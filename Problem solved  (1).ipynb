# 1.	Two coins are tossed 500 times, and we get:
Two heads: 105 times
One head: 275 times
No head: 120 times
Find the probability of each event to occur.


```python
two_heads = 105
one_head = 275
no_head = 120
total_tosses = 500
prob_two_heads = two_heads / total_tosses
prob_one_head = one_head / total_tosses
prob_no_head = no_head / total_tosses

# Print the probabilities
print(f"Probability of Two Heads: {prob_two_heads:.2f} or {prob_two_heads * 100:.2f}%")
print(f"Probability of One Head: {prob_one_head:.2f} or {prob_one_head * 100:.2f}%")
print(f"Probability of No Head: {prob_no_head:.2f} or {prob_no_head * 100:.2f}%")
```

# 5. A coin is thrown 3 times .what is the probability that atleast one head is obtained?

```python
prob_tails_single_toss = 1 / 2

# Probability of getting tails in all three tosses
prob_tails_all_three_tosses = prob_tails_single_toss ** 3

# Probability of getting at least one head
prob_at_least_one_head = 1 - prob_tails_all_three_tosses

# Print the probability
print(f"Probability of getting at least one head in three tosses: {prob_at_least_one_head:.2f} or {prob_at_least_one_head * 100:.2f}%")
```

 # 6.	In an interview the probability of Praveen to got selected is 0.4. The probability of Geetha to got selected is 0.5. The probability of Sam to got selected is 0.6. The probability of Suresh to got selected is 0.8. What is the probability that at least 2 of them got selected on that day?

```python
# Individual probabilities
prob_praveen_selected = 0.4
prob_geetha_selected = 0.5
prob_sam_selected = 0.6
prob_suresh_selected = 0.8

# Complementary probability of fewer than two selected
prob_fewer_than_two_selected = (1 - prob_praveen_selected) * (1 - prob_geetha_selected) * (1 - prob_sam_selected) * (1 - prob_suresh_selected)

# Probability of at least two selected
prob_at_least_two_selected = 1 - prob_fewer_than_two_selected

# Print the probability
print(f"Probability that at least two of them got selected: {prob_at_least_two_selected:.2f} or {prob_at_least_two_selected * 100:.2f}%")
```

# 7.	Find the probability of getting a numbered card when a card is drawn from the pack of 52 cards.

```python
numbered_cards_per_suit = 9  # 2 through 10

number_of_suits = 4  # Hearts, Diamonds, Clubs, Spades

# Total number of numbered cards
total_numbered_cards = numbered_cards_per_suit * number_of_suits

# Total number of cards in the deck
total_cards_in_deck = 52

# Probability of drawing a numbered card
prob_numbered_card = total_numbered_cards / total_cards_in_deck

# Print the probability
print(f"Probability of drawing a numbered card: {prob_numbered_card:.2f} or {prob_numbered_card * 100:.2f}%")
```

# 8.	There are 5 green 7 red balls. Two balls are selected one by one without replacement. Find the probability that first is green and second is red.

```python
# Number of green balls
green_balls = 5

# Number of red balls
red_balls = 7

# Total number of balls initially
total_balls_initial = green_balls + red_balls

# Probability of the first ball being green
prob_first_green = green_balls / total_balls_initial

# Total number of balls after the first draw
remaining_total_balls = total_balls_initial - 1

# Probability of the second ball being red given that the first was green
prob_second_red_given_first_green = red_balls / remaining_total_balls

# Probability of first being green and second being red
prob_first_green_second_red = prob_first_green * prob_second_red_given_first_green

# Print the probability
print(f"Probability of first being green and second being red: {prob_first_green_second_red:.2f} or {prob_first_green_second_red * 100:.2f}%")
```

# 9.	Three dice are rolled together. What is the probability as getting at least one '4'?

```python
# Probability of not getting a '4' on a single die
prob_no_4_single_die = 5 / 6

# Probability of not getting a '4' on all three dice
prob_no_4_all_dice = prob_no_4_single_die ** 3

# Probability of getting at least one '4'
prob_at_least_one_4 = 1 - prob_no_4_all_dice

# Print the probability
print(f"Probability of getting at least one '4' when three dice are rolled: {prob_at_least_one_4:.2f} or {prob_at_least_one_4 * 100:.2f}%")
```

# 10.	A and B are two candidates seeking admission in a college. The probability that A is selected is 0.7 and the probability that exactly one of them is selected is 0.6. Find the probability that B is selected.

```python
# Given probabilities
prob_A_selected = 0.7
prob_exactly_one_selected = 0.6

# Probability of not selecting B
prob_not_B = 1 - prob_A_selected

# Solving for P(B)
prob_B_selected = (prob_exactly_one_selected - prob_A_selected * prob_not_B) / (1 - prob_not_B)

# Print the probability
print(f"Probability that B is selected: {prob_B_selected:.2f} or {prob_B_selected * 100:.2f}%")
```
